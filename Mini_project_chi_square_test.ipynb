# Chi-square distribution
Chi-squared distribution is a probability distribution that describes the distribution of the sum of squared standard normal variables.

It is a continuous probaility distribution that takes only positive values, and its shape depends on a parameter called the degrees of freedom, k.


The mean of a chi-square distribution is equal to its degress of freedon (k)
and the variance is 2k. the range is 0 to infinity.




321px-Chi-square_pdf.svg.png

Chi-square distribution can be used to test whether observed data follows an expected distribution or not. For example, a researcher may want to test whether the distribution of observed heights in a population follows a normal distribution. It is called a goodness of fit test.

Another example for using Chi-square distribution in real life is when a sciencist wants to test whether a certain gene is associated with a particular disease by comparing the observed genotype frequencies in affected individuals to the expected frequencies based on the population genotype frequencies. And this is called the independence test.

# chi-square distributions code in python

```
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt


#x-axis ranges from 0 to 20 with .001 steps
x = np.arange(0, 20, 0.001)

#define multiple Chi-square distributions
plt.plot(x, chi2.pdf(x, df=4), label='df: 4')
plt.plot(x, chi2.pdf(x, df=8), label='df: 8') 
plt.plot(x, chi2.pdf(x, df=12), label='df: 12') 

#add legend to plot
plt.legend()
```

# What is Chi-square test
Chi-squared test is a statistical test that helps us determine if there is a significant difference between observed and expected frequencies in a categorical dataset.



To perform a chi-square test, we first calculate a test statistic called the chi-squared statistic. This statistic is calculated by taking the sum of the squared differences between the observed and expected frequencies, divided by the expected frequencies.

Once we have calculated the chi-square statistic, we compare it to the critical value. The critical value is a value from a table or calculator that tells us whether our chi-square statistic is statistically significant or not. If our chi-squared statistic is greater than the critical value, we can reject the null hypothesis (which states that there is no significant relationship between the two categorical variables) and conclude that there is a significant relationship between the two variables.



# null hypothesis

A company wants to know if there is a significant relationship between gender and preferred method of payment among their customers. They conduct a survey among a sample of 500 customers and ask them their gender and preferred method of payment (credit card, debit card, or cash).

null hypothesis:
There is a significant relationship between gender and preferred method of payment.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Generate random survey data to simulate the survey results:
n = 500
gender = np.random.choice(['male', 'female'], size=n)
payment = np.random.choice(['credit card', 'debit card', 'cash'], size=n)
```

Next, we'll create a contingency table to summarize the data:

```
# Create contingency table
contingency_table = pd.crosstab(gender, payment)
```



We can use this contingency table to perform a chi-square test to determine if there is a significant relationship between gender and preferred method of payment:


```
from scipy.stats import chi2_contingency

# Perform chi-square test
statistic, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square test statistic: {statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# The chi2_contingency function from scipy.stats takes the contingency table as input and returns four values: the chi-square test statistic, 
# the p-value, the degrees of freedom, and the expected frequencies.
```

```
# Create a stacked bar chart to visualize the contingency table:
from scipy.stats import chi2
df = 2
n = 500
x = np.arange(0, 20, 0.001)
pdf = chi2.pdf(x, df=2)
plt.plot(x, pdf)
plt.title('Gender and Payment Method')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

```

Since the p-value is greater than our significance level of 0.05, we fail to reject the null hypothesis and conclude that there is not a significant relationship between gender and preferred method of payment.

The output shows the chi-square statistic, the p-value and the degrees of freedom followed by the expected counts.

As expected, given the high p-value, the test result does not detect a significant relationship between the variables.
